```python
import pandas as pd
```

```python
import numpy as np
```

## Leer datos

```python
df = pd.read_excel("Precios Brendt 2024-10.xls",sheet_name='Datos a Utilizar')
```

```python
df = df.drop(df.index[0])
```

```python
df.columns = ['Dia', 'Precio']
```

## Retornos logaritmicos

```python
df['Retornos logaritmicos'] = np.log(df['Precio'] / df['Precio'].shift(1))
```

```python
df
```

## Media diaria

```python
media_diaria = df['Retornos logaritmicos'].mean()
```

```python
media_diaria
```

## Volatilidad diaria

```python
volatilidad_diaria = df['Retornos logaritmicos'].std()
```

```python
volatilidad_diaria
```
